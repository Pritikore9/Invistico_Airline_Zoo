```python
data2=data1.drop(['species'],axis=1)
```

```python
data2
```

```python
data2['class_type'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
```

```python
data2['animal_name']=l.fit_transform(data2['animal_name'])
```

```python
data2['class_type']=l.fit_transform(data2['class_type'])
```

```python
data2
```

```python
x=data2.iloc[:,data2.columns!='class_type'].values
y=data2.iloc[:,1].values
```

```python
x
```

```python
y
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
r=RandomOverSampler()
x_data,y_data=r.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2,random_state=1)
```

```python
!pip install xgboost
```

```python
import xgboost as xgb
```

```python
model=xgb.XGBClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
acc=accuracy_score(y_test,y_pred)*100
acc
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb=GaussianNB()
nb.fit(x_train,y_train)
```

```python
y_pred=nb.predict(x_test)
y_pred
```
